<h2><b>Train Test Split

```python
from sklearn.model_selection import train_test_split
X = valo_df_noagent.iloc[:, :-1]
y = valo_df_noagent.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 108)
```

<h3><b>Cross Validation

```python
def cross_validation(model, X, y, cv = 5):
    scoring = ['f1', 'accuracy']
    results = cross_validate(estimator = model, X = X, y = y, cv = cv, scoring = scoring, return_train_score=True)
    return {"Training accuracy scores": results['train_accuracy'],
              "Mean Training accuracy": results['train_accuracy'].mean(),
              "Training f1 scores": results['train_f1'],
              "Mean Training f1": results['train_f1'].mean(),
              "Validation accuracy scores": results['test_accuracy'],
              "Mean Validation accuracy": results['test_accuracy'].mean(),
              "Validation f1 scores": results['test_f1'],
              "Mean Validation f1": results['test_f1'].mean(),
              }
```

```python
model_score = {}
# model_score = {model: ['Mean Training accuracy', 'Mean Training f1', 'Mean Validation accuracy', 'Mean Validation f1']}
```

<h2><b>Model Building

<h3><b>Logistic Regression

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
LR_model = LogisticRegression(solver = 'liblinear')
```

```python
%%time
LR_result = cross_validation(LR_model, X_scaled, y, 10)
model_score['LR'] = [LR_result['Mean Training accuracy'], LR_result['Mean Training f1'], LR_result['Mean Validation accuracy'], LR_result['Mean Validation f1']]
```

```python
train_scaler = MinMaxScaler()
X_train_scaled = train_scaler.fit_transform(X_train)
X_test_scaled = train_scaler.transform(X_test)
```

```python
LR_model.fit(X_train_scaled, y_train)
LR_pred = LR_model.predict(X_test_scaled)
LR_pred_train = LR_model.predict(X_train_scaled)
```

```python
print('f1_score:\t', f1_score(y_test, LR_pred))
print('Train_accuracy:\t', accuracy_score(y_train, LR_pred_train))
print('Test_accuracy:\t', accuracy_score(y_test, LR_pred))
```

```python
# Plotting Confusion Matrix
confusion_matrix_train = confusion_matrix(y_train, LR_pred_train)
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix_train)
cm_display.plot()
print(accuracy_score(y_train, LR_pred_train))
plt.title('Train Data CF')
plt.plot()
```

```python
confusion_matrix_test = confusion_matrix(y_test, LR_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix = confusion_matrix_test)
cm_display.plot()
print(accuracy_score(y_test, LR_pred))
plt.title('Test Data CF')
plt.plot()
```

```python
# Feature Importance after Logistic Regression
LR_model.fit(X_train_scaled, y_train)
f_imp_LR = LR_model.coef_[0]
plt.figure(figsize = (14, 5))
ax = sns.barplot(x = list(X_train.columns), y = f_imp_LR)
plt.xticks(rotation = 90)
plt.plot()
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f')
```

<h7>Using RandomSearchCV

```python
%%time
from sklearn.model_selection import RandomizedSearchCV

max_depth = [3, 4, 5, 6]
alpha = [0, 0.2, 0.3, 0.4, 0.5, 1, 1.5]
n_estimators = [60, 75, 100, 115]
min_child_weight = [5, 10, 20, 30]

param_grid = {'max_depth': max_depth,
              'alpha': alpha,
              'n_estimators': n_estimators,
              'min_child_weight': min_child_weight}

XGB_model = xgb.XGBClassifier()

XGB_grid = RandomizedSearchCV(estimator = XGB_model,
                       param_distributions = param_grid,
                       cv = 5,
                       verbose=0,
                       n_jobs = -1,
                       n_iter=40)

XGB_grid.fit(X, y)
```

```python
# Using soft voting of SVC with rbf kernel, LGB, SVC with linear kernel and LR
from sklearn.ensemble import VotingClassifier
SVC_model_rbf = SVC(kernel='rbf', probability=True)
estimators = [('SVC_rbf', SVC_model_rbf), ('LGB', LGB_model), ('SVC', SVC_model), ('LR', LR_model)]
vc_soft_model = VotingClassifier(estimators = estimators, voting = 'soft')
vc_soft_result = cross_validation(vc_soft_model, X_scaled, y, 5)
print('Mean Validation accuracy:\t', vc_soft_result['Mean Validation accuracy'])
```
